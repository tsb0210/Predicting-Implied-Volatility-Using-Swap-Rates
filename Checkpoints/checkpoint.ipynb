```python
def annuity(tenor, expiry,forward_array):
    annuity=0
    annuity_list = np.zeros(len(forward_array))
    for i in range(1,(tenor*2)+1):
    
        annuity_list += 0.5 * np.exp(-0.5*( expiry+ i) * (forward_array/10000))
    return annuity_list
        
```

```python
def bachelier_call(strike_premium,T,sigma,A):
    
    d1=strike_premium/ (sigma* math.sqrt(T))
    P= A*sigma*math.sqrt(T)*((d1*norm.cdf(d1))+norm.pdf(d1))    
    return P     
```

```python
swap_1yr_merged["1m1yr Premium"]=bachelier_call(0,1/12,swap_1yr_merged['1m Expiry']/10000,swap_1yr_merged['1m1yr Annuity'])*10000
swap_1yr_merged["3m1yr Premium"]=bachelier_call(0,3/12,swap_1yr_merged['3m Expiry']/10000,swap_1yr_merged['3m1yr Annuity'])*10000
swap_1yr_merged["6m1yr Premium"]=bachelier_call(0,1/2,swap_1yr_merged['6m Expiry']/10000,swap_1yr_merged['6m1yr Annuity'])*10000
swap_1yr_merged["1yr1yr Premium"]=bachelier_call(0,1,swap_1yr_merged['1yr Expiry']/10000,swap_1yr_merged['1yr1yr Annuity'])*10000
swap_1yr_merged["2yr1yr Premium"]=bachelier_call(0,2,swap_1yr_merged['2yr Expiry']/10000,swap_1yr_merged['2yr1yr Annuity'])*10000
swap_1yr_merged["5yr1yr Premium"]=bachelier_call(0,5,swap_1yr_merged['5yr Expiry']/10000,swap_1yr_merged['5yr1yr Annuity'])*10000


```

```python
def get_implied_vol_bs(S_0,K,T,r,observed_px, initial_guess): 
    root_fn = lambda x: (bs_call(S_0,K,T,x,r) - observed_px)
    return root(root_fn ,initial_guess)['x'][0]
```

```python
def bs_call(S_0,K,T,sigma,r):
    d_1 = (np.log(S_0/K) + T*(r + (sigma**2)/2))/(sigma*np.sqrt(T))
    d_2 = d_1 - sigma*np.sqrt(T)
    return S_0*norm.cdf(d_1) - K*np.exp(-r*T)*norm.cdf(d_2)
```
